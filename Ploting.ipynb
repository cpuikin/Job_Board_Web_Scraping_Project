```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Final_Data.csv')
df = df.drop('Unnamed: 0',axis=1)
df = df.drop('Unnamed: 0.1',axis=1)
df = df.drop('index',axis=1)
df
```

```python
df_skill = df.loc[:, ['Python','SQL','Machine Learning','Spark','Java','Data Mining','Hadoop','Tableau','Hive','SAS','Big Data','GCP','AI','R','AWS','Azure']]
df_skill
```

```python
df001 = pd.DataFrame(df_skill.sum())
df001 = df001.reset_index()
df001.columns = ('Skill','Count')
df001 = df001.sort_values(by='Count', ascending=False)
df001
```

```python
sns.set(color_codes=True)
mpl.rcParams['figure.figsize'] = [5,5]
sns.barplot(data=df001,y='Skill', x='Count')
plt.title('Most Skills Company Need')
plt.show()
```
